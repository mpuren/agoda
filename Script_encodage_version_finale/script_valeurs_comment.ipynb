```python
# script qui ouvre un fichier json et récupère les valeurs des comments

import os
import json

dossier_json = "json_data"

for fichier_json in os.listdir(dossier_json):
    if fichier_json.endswith(".json"):
        chemin_json = os.path.join(dossier_json, fichier_json)
        with open(chemin_json, "r", encoding="utf-8") as f:
            data = json.load(f)
            print("Fichier:", fichier_json)
            for item in data:
                if "comment" in item:
                    print(item["comment"])
                else:
                    pass
            print("--------------------")
```

```python
import os
import json

dossier_json = "json_data"
commentaires_uniques = set()

for fichier_json in os.listdir(dossier_json):
    if fichier_json.endswith(".json"):
        chemin_json = os.path.join(dossier_json, fichier_json)
        with open(chemin_json, "r", encoding="utf-8") as f:
            data = json.load(f)
            print("Fichier:", fichier_json)
            for item in data:
                if "comment" in item:
                    commentaire = item["comment"]
                    commentaires_uniques.add(commentaire)

            print("Commentaires uniques :")
            for commentaire in commentaires_uniques:
                print(commentaire)
            print("--------------------")
```

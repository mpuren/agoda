```python
from lxml import etree
import os
import re

# Obtenez la liste des fichiers dans le dossier "xml_data"
xml_files = os.listdir("C:/Users/brune/Desktop/stage_AGODA/xml_data")

# Utiliser une expression régulière pour extraire le numéro de page à partir du nom de fichier
# Triez les fichiers en se basant sur l'élément de fin du nom des fichiers
page_number_pattern = re.compile(r'^.*_p(\d+)\.xml$')
xml_files_sorted = sorted(xml_files, key=lambda x: int(page_number_pattern.match(x).group(1)))

# Ouvrir le premier fichier XML (la page principale)
with open(os.path.join('xml_data', xml_files_sorted[0]), 'r') as f:
    root = etree.fromstring(f.read().encode('utf-8'))
    
# Obtenez l'élément "body" de la racine du premier fichier XML
body_element = root.find('.//{http://www.tei-c.org/ns/1.0}body')

# Pour chaque fichier XML suivant dans le dossier "xml_data", extraire les éléments <text> et <body> de la racine de ce fichier XML et les ajouter en tant que sous-éléments de la racine du premier fichier XML (la page principale)
for filename in xml_files_sorted[1:]:
    with open(os.path.join('xml_data', filename), 'r', encoding='utf-8') as f:
        page = etree.fromstring(f.read().encode('utf-8'))
    for child in page.iter('{http://www.tei-c.org/ns/1.0}body/*'):
        # Ajouter un commentaire indiquant que ce fichier a été ajouté en tant que nouvelle page
        comment = etree.Comment(f"New page added from {filename}")
        body_element.append(comment)
        for child in page.find('.//{http://www.tei-c.org/ns/1.0}body').getchildren():
            body_element.append(child)

# Enregistrer le fichier XML final
with open('C:/Users/brune/Desktop/stage_AGODA/final_doc/final.xml', 'wb') as f:
    f.write(etree.tostring(root, encoding="utf-8", pretty_print=True))
```
